### 1.2 GropuAの国だけを抽出

```python
#  GroupAの国（ロシア、サウジアラビア、エジプト、ウルグアイ）だけ抽出
GroupA = soccer[(soccer['Team'] == 'Russia') | (soccer['Team'] == 'Saudi Arabia') | (soccer['Team'] == 'Egypt') | (soccer['Team'] == 'Uruguay')]
```

##  1.3  グループステージのデータだけを見る

```python
GroupA = GroupA[GroupA['Round'] == 'Group Stage']
```

```python
GroupA
```

```python
## 可視化
```

```python
# 国ごとのシュート数
plt.figure(figsize = (30,15))
# 国ごとにまとめる
A_mean = GroupA.groupby('Team').mean()
# 国の順序を最終成績順に並び替え
A_mean = A_mean.loc[['Uruguay','Russia','Saudi Arabia','Egypt']]
x = np.arange(0,4)

plt.bar(x+0.2,A_mean['Attempts'],width=0.2)
plt.bar(x,A_mean['On-Target'],width=0.2)
plt.bar(x-0.2,A_mean['Off-Target'],width=0.2)
plt.title('国ごとのシュート数',fontsize = 30)
plt.xticks(x,A_mean.index,fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('国名',fontsize = 20)
plt.ylabel('平均シュート数',fontsize=20)
plt.legend(['合計シュート数','シュート数枠内','シュート数枠外'],fontsize=20)
plt.show()
```

```python
# 国ごとのシュート数
plt.figure(figsize = (30,15))


plt.bar(x,A_mean['Saves'])
plt.title('国ごとの平均セーブ数',fontsize = 30)
plt.xticks(x,A_mean.index,fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('国名',fontsize = 20)
plt.ylabel('平均セーブ数',fontsize=20)
plt.show()
```
